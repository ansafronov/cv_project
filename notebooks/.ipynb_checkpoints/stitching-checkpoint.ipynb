```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
from tqdm import tqdm
```

## Read data

```python
sample = 6

imgs_path = os.listdir(f'../data/cv_imgs/{sample}')

imgs = []
for img in sorted(imgs_path):
    to_append = {}
    to_append['img'] = cv2.imread(f'../data/cv_imgs/{sample}/{img}')
    to_append['img_gray'] = cv2.cvtColor(to_append['img'], cv2.COLOR_BGR2GRAY)
    imgs.append(to_append)
```

```python
imgs = [
    {
        'img': cv2.imread('../data/img_3_0.jpg')
    }, 
    {
        'img': cv2.imread('../data/img_3_1.jpg')
    }
]
```

```python
sift = cv2.SIFT_create()

for img in imgs:
    kpts, desc = sift.detectAndCompute(img['img_gray'], None)
    img['keypoints'] = kpts
    img['descriptors'] = desc
```

```python
KNN = 2
LOWE = 0.7
TREES = 5
CHECKS = 50

matcher = cv2.FlannBasedMatcher({'algorithm': 0, 'trees': TREES}, {'checks': CHECKS})
```

```python
cur_img = imgs[0]
for img in tqdm(imgs[1:]):
    matches = matcher.knnMatch(cur_img['descriptors'], img['descriptors'], k=KNN)

    positive = []
    for left_match, right_match in matches:
        if left_match.distance < LOWE * right_match.distance:
            positive.append(left_match)

    source_keyponts = np.array([cur_img['keypoints'][m.queryIdx].pt for m in positive])
    dest_keypoints = np.array([img['keypoints'][m.trainIdx].pt for m in positive])
    
    M, mask = cv2.findHomography(
        source_keyponts,
        dest_keypoints,
        cv2.RANSAC,
        ransacReprojThreshold=.5,
        maxIters=20000,
        confidence=0.995
    )

    new_img = cv2.warpPerspective(
        cur_img['img_gray'], M, (cur_img['img_gray'].shape[1] + img['img_gray'].shape[1], cur_img['img_gray'].shape[0])
    )
    new_img[0:cur_img['img_gray'].shape[0], 0:cur_img['img_gray'].shape[1]] = cur_img['img_gray']

    cur_img['img_gray'] = new_img
    kpts, desc = sift.detectAndCompute(cur_img['img_gray'], None)
    cur_img['keypoints'] = kpts
    cur_img['descriptors'] = desc
```

```python
plt.figure(figsize=(15, 8))
plt.imshow(cur_img['img_gray'], cmap='gray')
plt.show()
```

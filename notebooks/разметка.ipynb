```python
import matplotlib.pyplot as plt
import os
import json
import cv2
```

```python
img_number = 1
```

```python
imgs_path = os.listdir(f'data/{img_number}')

len(imgs_path)
```

```python
idx = 0
```

```python
for i, img_path in enumerate(imgs_path):
    if i == idx:
        image = cv2.imread(f'data/{img_number}/{img_path}')
        
        plt.figure(figsize=(15, 8))
        plt.imshow(image, cmap='gray')

        points_raw = [
            {"x": 100, "y": 200},
            {"x": 100, "y": 300},
            {"x": 100, "y": 400},
            {"x": 100, "y": 500},
            {"x": 100, "y": 600},
            {"x": 100, "y": 700},
        ]

        for point in points_raw:
            plt.plot(
                point['x'],
                point['y'],
                'r.'
            )

        plt.show()
        
        points = []
        for p in points_raw:
            points.append({
                "x": p["x"] / image.shape[1],
                "y": p["y"] / image.shape[0]
            })
            
        with open(f'data/{img_number}/{img_path}.json', 'w') as f:
            json.dumps(points)
```
